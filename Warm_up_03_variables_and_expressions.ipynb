```python
print(f"The value is  {int(v_car)}  and the type is  {type(int(v_car))}")
v_car_int = int(v_car)
```

**Now that you changed the type, are the two variables equal? Discuss why or why not by using a comment after the calculation with the symbol #**

```python
print(v_car == v_car_int) # They are not equal because v_car is a float with decimal places, and v_car_int is an integer, so the values are different.
```
